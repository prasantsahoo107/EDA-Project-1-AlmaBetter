# **3. Data Wrangling**

## Data Wrangling Code

Data Cleaning


>  1.1. Duplicate Rows are removed


Data pre-processing


> 2.1. Assigning values to cells with missing values

 >2.2. Changing values to appropriate data types

 
 >2.3. Additional columns added as per requiremennt


```python
ds1.drop_duplicates(inplace = True)    #duplicate rows are removed
```

```python
ds1.shape    #after removing duplicate values
```

```python
ds1[['company','agent']] = ds1[['company','agent']].fillna(0)    #We will replace null values by 0 in these columns
```

```python
ds1['children'].fillna(ds1['children'].mean(), inplace = True)   #We will replace the null values under this column with mean value of children
```

```python
ds1['country'].fillna('others', inplace = True)                  #We will replace the null values under this column with 'others' string
```

```python
ds1.isnull().sum().sort_values(ascending = False)[:5]            #Confirming if null values are replaced
```

```python
ds1[ds1['adults']+ds1['babies']+ds1['children'] == 0].shape       
ds1.drop(ds1[ds1['adults']+ds1['babies']+ds1['children'] == 0].index, inplace = True)     #Rows with total number of adults, children or babies equal to zero are removed
```

```python
 # Converting datatype of columns 'children', 'company' and 'agent' from float to int.
ds1[['children', 'company', 'agent']] = ds1[['children', 'company', 'agent']].astype('int64') 
```

```python
# Changing datatype of column 'reservation_status_date' to data_type.
ds1['reservation_status_date'] = pd.to_datetime(ds1['reservation_status_date'], format = '%Y-%m-%d')
```

```python
# Adding total staying days in hotels 
ds1['total_stay'] = ds1['stays_in_weekend_nights']+ds1['stays_in_week_nights']
```

```python
# Adding total people num as column, i.e. total people num = num of adults + children + babies
ds1['total_people'] = ds1['adults']+ds1['children']+ds1['babies']
```

## Chart 10

```python
waiting_bookings = ds1[ds1['days_in_waiting_list'] !=0]  # Selecting bookings with non zero waiting time


fig, axes = plt.subplots(1, 2, figsize=(18, 8))
sns.kdeplot(ax=axes[0],x = 'days_in_waiting_list', hue = 'is_canceled' , data = waiting_bookings)
sns.kdeplot(ax = axes[1], x = ds1['lead_time'], hue = ds1['is_canceled'])
plt.show()
```

Why did you pick the specific chart?

Kernel Density Plots (KDE Plots) are a useful tool for exploring the distribution of a continuous variable. They provide a smooth estimate of the underlying probability density function of the variable and can help you identify patterns and trends in the data

What is/are the insight(s) found from the chart?

While it is evident that the majority of cancelled reservations have waiting periods under 150 days, the majority of confirmed reservations also do. This demonstrates that the waiting period has no bearing on booking cancellations.

Additionally, because the cancellation and non-cancellation curves for lead time are comparable, lead time has no bearing on booking cancellation.
